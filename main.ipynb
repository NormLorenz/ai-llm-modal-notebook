### ğŸ§ª Step 3: Access the Files in Your Modal App

In your Python code:

```python
import modal

app = modal.App()
volume = modal.NetworkFileSystem.persisted("my-volume")

@app.function(network_file_systems={"/data": volume})
def run():
    with open("/data/my_project/some_file.txt") as f:
        print(f.read())
```

You can now read any file you uploaded using the `/data` path.
